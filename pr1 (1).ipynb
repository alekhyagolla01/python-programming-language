```python
class CAL:
  def __init__(self,v1,v2):
    self.v1 = v1
    self.v2 = v2
  def addtion(self):
    print(self.v1 + self.v2)
  def subtraction(self):
    print(self.v1 - self.v2)
  def multiplication():
    print(self.v1 * self.v2)
  def division():
    print(self.v1 / self.v2)
va1 = int(input(f'enter'))
va2 = int(input(f'enter'))
op = input('enter')
anil = CAL(va1,va2)
if op == '+':
  anil.addtion()
elif op == '-':
  anil.subtraction()
elif op == '*':
  anil.multiplication()
else:
  anil.division()
```

```python
class V:
  def __init__(self,a,m):
    self.a = a
    self.m = m
  def add(self):
    self.re = self.a + self.m
  def sub(self):
    self.res = self.a -self.m
  def common(self):
    print(self.re / self.res)
anil = V(12,56)
anil.add()
anil.sub()
anil.common()

```
