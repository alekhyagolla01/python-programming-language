```python
n = 6
s = 0

for i in range(1,n):
  if n % i == 0:
    s = s + i

if n == s:
  print(f"{n} : is Perfect Number")
else:
  print(f"{n} : is Not Perfect Number")
```

```python
def sharuk(n):
  s = 0

  for i in range(1,n):
    if n % i == 0:
      s = s + i

  if n == s:
    print(f"{n} : is Perfect Number")
  else:
    print(f"{n} : is Not Perfect Number")


sharuk(6)
```
