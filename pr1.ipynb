```python
class PRIME_NUMBER:
  try:
    n = 7
    def sharuk(self):  # self -> anil only
      t = 0
      for i in range(1,self.n+1):
        if self.n % i == 0:
          t = t + 1

      if t == 2:
        print(f"{self.n} : Is a Prime Number")
      else:
        print(f"{self.n} : Is not a Prime Number")

  except Exception as e:
    er_t , er_m , er_line = sys.exc_info()
    print(f"Error in Line no : {er_line.tb_lineno} Due to : {er_t} : reason : {er_m}")


anil = PRIME_NUMBER()
anil.sharuk()
```
