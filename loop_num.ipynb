```python
# largest digit in number
a = int(input('enter the number'))
largest = 0
while a > 0:
  s = a % 10
  if s > largest:
    largest = s
  a = a//10
print(largest)
```

```python
a = int(input('enter the number'))
smallest = 1
while a <= 0:  #smallest number
  s = a % 10
  if s < smallest:
    smallest = s
    a = a //10
print(smallest)
```

```python
a = 'banana' #frequency
b = {}
for i in a:
  if i in b:
    b[i] = b[i] +1
  else:
    b[i] = 1
print(b)
```

```python
a = 123 # sum digits
sum = 0
while a > 0:
  s = a % 10
  sum = sum + s
  a = a//10
print(sum)
```

```python
a = 123 # product
product = 1
while a > 0:
  s = a % 10
  product = product * s
  a = a//10
print(product)
```

```python
a = 121
t = a
s = 0
while a > 0: # palindrome
  res = a % 10
  s = s*10 +res
  a = a //10
if t == s:
 print('it is a palindrome')
else:
 print('it is not palindrome')
```

```python
a = 121
t = a
s = 0 #amstrong
while a > 0:
  res = a % 10
  s = s + res * res *res
  a = a //10
if t == s:
 print('it is amstrong')
else:
 print('it is not not amstrong')
```

```python
n = 36
for i in range(1,n+1):
  if i * i == n:
    print('ps') # perfect square
    break
else:
    print('nps')
```

```python
n = '12345'
count = 0 # even count

for i in n:
  if int(i) % 2 == 0:
    count = count + 1

print(count)
```
